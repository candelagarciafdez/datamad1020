#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

#### Whether the bin size affects your plot? Why?

Sí, cuanto más ancho es el binado, se suaviza la forma de la distribución principal, pero se pierde información de estructuras internas. Asímismo, con bins más estrechos se ve más ruido.

#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here
sns.heatmap(titanic.corr(), square = True, annot=True, cmap = sns.diverging_palette(120,20, as_cmap = True), vmax = 1., vmin = -1.)
```

#### According to the previous plot, what are the most correlated feature? Comment.

`Pclass` y `Fare` muestran un coeficiente de correlación de Pearson más alto, aunque en negativo, por lo que a mayor `Pclass`, menos `Fare` (se entiende si hace referencia ordinal a las clases de billete).
